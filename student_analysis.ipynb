1. LOAD DATASET


* Use pandas to read student-mat.csv dataset

```python
df=pd.read_csv("/student-mat.csv")
```

read csv file using pandas

2. EXPLORE & CLEAN DATA

* Check missing values (.isnull()).

```python
missing_values=df.isnull().sum()
print(missing_values)
```

* Remove duplicates.

```python
duplicates=df.duplicated().sum()
print(duplicates)
```

check duplicates

```python
df.drop_duplicates(inplace=True)
```

drop duplicates

* Inspect dataset shape & dtypes.

```python
df.shape
```

```python
print(df.dtypes)
```

3. Analysis Questions

• Average final grade (G3).

```python
print("Average Final Grade:",df['G3'].mean())
```

using mean function

• How many students scored above 15?

```python
print("Number of Students with scored Above 15:",(df['G3']>15).sum())
```

• Is study time correlated with performance?

```python
print("Correlation between Study Time and Performance:",df['studytime'].corr(df['G3']))
```

using co-relation function

Here correlation between study time and performance is 0.097819, it is very close to 0.
So we can say that, there is almost no linear relationship between study time and performance.

• Which gender performs better on average?

```python
better_gender=df.groupby('sex')['G3'].mean()
print(better_gender)
```

Here male average is higher then females, so male perform better in average.
